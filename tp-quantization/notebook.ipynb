```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def quantize(float_value, min_range, max_range, zero=0):
    b = 8
    S = (max_range - min_range) / (2**b - 1)
    q = (float_value / S).astype(np.int8)
    return q

def to_float(int_values, min_range, max_range):
    b = 8
    S = (max_range - min_range) / (2**b - 1)
    return int_values.astype(np.float32) * S
```

```python
def display_error_histogram(values, min_range, max_range):
    q = quantize(values, min_range, max_range)
    dq = to_float(q, min_range, max_range)
    errors = dq - values

    plt.hist(errors, bins=30)
    plt.title("Histogramme des erreurs de quantization")
    plt.xlabel("Erreur")
    plt.ylabel("Fréquence")
    plt.show()
```

```python
def quantization_error_stats(values, min_range, max_range):
    q = quantize(values, min_range, max_range)
    dq = to_float(q, min_range, max_range)
    errors = dq - values

    return errors.mean(), errors.std()
```

```python
values_1 = np.random.uniform(0, 0.5, 1000)

display_error_histogram(values_1, 0, 1)

mean_1, std_1 = quantization_error_stats(values_1, 0, 1)
print("Situation 1 — mean:", mean_1, "std:", std_1)
```

```python
def quantize_bad(float_value):
    S = 1e-12
    return (float_value / S).astype(np.int8)

def to_float_bad(int_values):
    S = 1e-12
    return int_values.astype(np.float32) * S

values_1bis = np.random.uniform(0, 1, 1000)

q_bad = quantize_bad(values_1bis)
dq_bad = to_float_bad(q_bad)
errors_bad = dq_bad - values_1bis

plt.hist(errors_bad, bins=30)
plt.title("Histogramme des erreurs — S ≈ 0 (mauvais)")
plt.show()

print("Mean bad:", errors_bad.mean(), "Std bad:", errors_bad.std())
```

```python
display_error_histogram(values_1bis, 0, 1)
mean_opt, std_opt = quantization_error_stats(values_1bis, 0, 1)
print("Mean optimal:", mean_opt, "Std optimal:", std_opt)
```
